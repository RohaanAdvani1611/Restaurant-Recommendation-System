## 6. Data Visualization with different types of charts for analyzing and presenting the data

### a)      Handling the outliers in the data, if any , using a Boxplot.

```python
eda = eda3
```

```python
eda.boxplot(column=['Pricefortwo'])
```

```python
IQR_Pricefortwo=eda['Pricefortwo'].quantile(0.75)-eda['Pricefortwo'].quantile(0.25)
print(IQR_Pricefortwo)
```

```python
Upper_OutlierLimit_Pricefortwo = eda['Pricefortwo'].quantile(0.75) + 1.5*IQR_Pricefortwo
Lower_OutlierLimit_Pricefortwo = eda['Pricefortwo'].quantile(0.25) - 1.5*IQR_Pricefortwo
print(Upper_OutlierLimit_Pricefortwo)
print(Lower_OutlierLimit_Pricefortwo)
```

```python
OutlierValues_Pricefortwo = eda[(eda['Pricefortwo']>=Upper_OutlierLimit_Pricefortwo) | (eda['Pricefortwo']<=Lower_OutlierLimit_Pricefortwo)]
```

```python
OutlierValues_Pricefortwo
```

```python
eda.boxplot(column=['Pricefortwo_scaled'])
```

```python
IQR_Pricefortwo_scaled=eda['Pricefortwo_scaled'].quantile(0.75)-eda['Pricefortwo_scaled'].quantile(0.25)
Upper_OutlierLimit_Pricefortwo_scaled = eda['Pricefortwo_scaled'].quantile(0.75) + 1.5*IQR_Pricefortwo_scaled
Lower_OutlierLimit_Pricefortwo_scaled = eda['Pricefortwo_scaled'].quantile(0.25) - 1.5*IQR_Pricefortwo_scaled
print(Upper_OutlierLimit_Pricefortwo_scaled)
print(Lower_OutlierLimit_Pricefortwo_scaled)
```

```python
for i in range(0,7245):
    if eda['Pricefortwo'][i] >= Upper_OutlierLimit_Pricefortwo:
        eda['Pricefortwo'][i] = Upper_OutlierLimit_Pricefortwo
    elif eda['Pricefortwo'][i] <= Lower_OutlierLimit_Pricefortwo:
        eda['Pricefortwo'][i] = Lower_OutlierLimit_Pricefortwo
    else:     
        continue
eda.boxplot(column=['Pricefortwo'])
```

```python
for i in range(0,7245):
    if eda['Pricefortwo_scaled'][i] >= Upper_OutlierLimit_Pricefortwo_scaled:
        eda['Pricefortwo_scaled'][i] = Upper_OutlierLimit_Pricefortwo_scaled
    elif eda['Pricefortwo_scaled'][i] <= Lower_OutlierLimit_Pricefortwo_scaled:
        eda['Pricefortwo_scaled'][i] = Lower_OutlierLimit_Pricefortwo_scaled
    else:     
        continue
eda.boxplot(column=['Pricefortwo_scaled'])
```
